## 1. Give some examples of some random sampling techniques.

When we need to anlayse a certain population,It is not possible to collect data for whole population as it is very time consuming and costly.So, what we do is we draw samples from population in such a way that sample can be a good representation of population.There are two types of sampling methods-Random and non random samplings.In question ,Its asked to describe random sampling methods.
There are many types of sampling methods.I am listing here 4 main types of sampling methods used -

**1)Simple Random sampling-** In this method,each member as equal chance(probability) of getting selected.There is no bias.This method is very simple.But, this method is expensive and is not possible when there isn't any group to select sample from.Example-Giving a group of people a random number and then using a random number generater to generate numbers.Select the person with generated number.

**2)Stratified Random sampling-** Population may consist of many groups and if all groups need to be equally represented in sample,then we use stratified sampling.For ex-To prdeict election results from exit polls there needs to be equal representation of population  in sample selected for exit polls for prediction to be correct.But this requires another sampling method to select from sample each group of population.

**3)Systematic sampling-** Every data point is selected after a certain distance or interval from a starting point.First datapoint is selected at random to avoid bias.For example-To select a sample of bottles to test quality we select first bottle at random and then select every 40th bottle coming out of production line  following first bottle.

**4)Cluster sampling-** This method divides population into subgroups and then selects sample from  each subgroup using random sampling.For example-To know the literacy rate,A district is divided into smaller groups like towns or villages and samples are selected from towns or villages.
