## 1. What is the curse of dimensionality? Can you list some ways to deal with it?

Curse of dimensionality implies as number of dimension increases the distance between points in n dimensional space increases and the space becomes sparse. Sparse space leads to an exponential incraese in number of possible unique rows which makes it tough for our model to generalize.We can understand this w.r.t to 2d example that if we have 3 points and we want to capture the pattern,It's very tough.Thus,to improve our model performance, increase in number of dimensions of our data  leads to the need of more amount of datapoints . As number of dimensions increases we need more number of points for points to be equidistant as compared to number of points in low dimension for same distance.
For example-
If we have 3 points in 1 line or dimension equidistant we would need a matrix of 9 points in 2 dimension to maintain the same sparsity.

## 2. How do L1 and L2 regularization differ in improving the accuracy of machine learning models? Implications of using each technique and which technique is commonly preferred to boost the model’s accuracy rate and why?


Regularization is used to avoid model overfitting.These methods avoid overfitting by panelizing the regression coefficients of high valueThere are two methods used for regularization-

** 1) Lasso **- Lasso refers to Least Absolute Shrinkage and Selection Operator.Absolute Shrinkage here denotes that via regularization we try to reduce the magnitude of coefficients by adding the absolute value of coefficients multiplied by a hyperparameter(lambda) to the loss function.Larger the value of lambda more will be the shrinkage.This helps in reducing model complexity which in turns reduces overfitting.Selection operator term in lasso represents that this can also be used to reduce the number of features as it has ability to shrink some less important feature coefficients to zero.

** 3) Ridge- **   In ridge regression instead of adding absolute value to coefficients we add square of coefficients multiplied by a hyperparameter(lambda).

Lasso regularization attempts to estimate the median of data, ridge regularization makes estimation for the mean of the data in order to evade overfitting.Since, lasso regression uses median of data it is robust to ouliers whereas ridge isn't as it uses mean of data.

L1 regularization has a sparse solution that's why it is used for feature selection whereas ridge doesn't gives sparse solution.

In lasso, since contour of RSS can cut the absolute coefficient curve that is a diamond shape at more than one point ,Thus giving more that 1 solution while in case of ridge regression contour of RSS can cut square of weight curve that is circle at only one point giving only one solution.

Ridge regularization is preferred when model accuracy is concerned as it is able to learn complex data patterns and is more stable when features are correlated to each other. Ridge  regression performs similar to unregularized regression when features are correlated and coefficients fluctuate alot even for small changes.

Lasso regression is computationally inefficient because of sparse conditions as compared to ridge regression.Sparsity leads to computational inefficiency.

## 3. What is stratified cross-validation and when should we
## use it?

Stratified cross validation refers to the CV technique in which when training dataset is divided into train and validation set in such a way that the proportion of dependent feature is same in both sets as in original dataset.For ex - if original dataset has 100 datapoints with 20 belonging to class '0' and 80 belonging to class '1'.We keep 80 datapoints for training and 20 for validation.Out of 80 training data points 16 datapoints will be of class '0'  and 64 will be of class '1'.While in validation set 16 points will be of class '1' and 4 will be of class '0'.

We use stratified cross validation when there is high imbalance in classes of dependent feature and we have less datapoints.We have only 1000 datapoints and out of these 900 are of one class and only 100 for second class. In this case its better to use startified CV so that we can have sufficient second class datapoints when  training our model.

## 4. Briefly explain Hyperparameter Optimization strategies and optimization strategy performance.

There are 3 famous hyperparameter optimization strategies used in machine learing-

**1)Grid Search CV -**

In this technique every possible combination of hyperparameters are used for training.The combination which is giving best result can be choosen.This is very time taking and computationally exhaustive.It is preferred when we have small dataset.

**2)Random Search CV-** 

In this startegy, Based on x number of iterations set by us,randomly hyperparameters are combined x times and used for training .The best combination is choosen from x combinations based on training accuracy.This saves a lot of time as compared to grid search CV when dealing with large datasets.This strategy is extensively used in deep learning.

**3)Bayesian search CV-**

This works on the priciple of bayes theoram.It learns from the random combination of hyperparameters based on its performace and then uses that learning to determine next combination.Based on these steps,Bayesian serch will come up with the best combination of hyperparameters.Advantage of this strategy over random search cv is that this doesn't combines hyperparameters randomly and overgridsearch cv is that it doesn't go through every possible combination.This is a probablistic approach.
