## 2. What are the differences between Adaboost and Gradient Boosting?

image.png

Adaboost and gradient boost both are boosting techniques  in which multiple trees are used in sequence i.e boost the performance of next tree using errors of previous tree.Although both are boosting techniques but they differ in their way of working-

**1) Basic difference-**

Adaboost uses a stump(a node and two leaves) as base learner and combines the result of multiple weak learners to come up with a strong learner.The more error a stump makes in preiction the lower say it will have in final prediction.On the other hand, gradient boosting uses a full grown tree(except for only first tree which is only a leaf).The pseudo residuals(difference between prediction and actual values) are as data to grow trees.A Learning rate is used to prevent overfitting.GBM uses gradient descent algorithm to optimize loss function.

**2)Loss function-** Exponential loss function is used for adaboost while gradient boost uses MSE for regression problem and binary cross entropy for classification.

**3)Flexibility-** Adaboost was the first boosting algorithm designed for mainly binary classification and  uses only one loss funtion and is less flexible while GBM is more flexible and has different approaches for both classification and regression.




**4) Approach-**

Ada boost increases the weight of the incorrectly classified sample by the stump,So that the stump gets more penalized when it misclassifies this sample again while GBM uses gradient descent algorithm to minimize loss function.

## 3. What is Model Explainability and why is it required?


There are certain fields like healthcare ,credit lend  where only model performance isn't enough. Alongwith model performance which features are affecting the performance and weather their impact is poitive or negative are also important to be determined.This is where model explainability comes into picture.

**Why?**

**1)Accountability-**  When our model performance makes wrong decision negatively,it is important to know what caused the model to know what caused it so that it can be avoided in future.

**2)Trust-**  Explainability is  crucial to building trust.Trust becomes important in high risk domains like healthcare and credit lending.Before deployment of model, all stakeholders must fully understand what the model does.

**3)Performance-** If we know how different features affect our model,performance can also be increased.We can determine what's causing error and can rectify it to improve performance.


## 4. What is the basic difference between ML Interpretability and ML Explainability?

image.png

For AI/ML methods, the terms interpretability and explainability are commonly interchangeable.We know that if a ball is let go on a slope ,it will roll down and its speed will increase.This in interpreatability but why this is happening is explaiability.

Interpretibilbity is when one wants to understand the inner mechanics of problem or to attach a cause to an effect.Interpretable refers to  tranparency i.e to know all weights and biases. It is the extent to which you are able to predict what is going to happen, given a change in input or algorithmic parameters.But interpretablilty comes at the cost of performance.

Explainabilty is understanding the effect of features on the model output or  explain the behavior of model in human terms.

Linear regression,decision trees,Logistic regression have good interpretability but their accuracy isn't up to the mark .On the other hand ,ensemble models,Deep learning models are best when it comes to accuracy but they aren't interpretable.So,there is a kind of inverse relation between interpretability and accuracy.The reason being deep learning projects data to higher dimensions for better performance and with increase in dimension ,interpretabilty reduces.But if we want both performance and have general understanding of model then explainabilty is the way.
