## 1. a. How are confidence tests and hypothesis tests similar? How are they different?

## b. What is the difference between type 1 error and type 2 error?


**Difference-**

In hypothesis test, we assume a population parameter value and then based on a sample we test whether the the population parameter we assume(made a hypothesis) is correct or not based on test statistic value.On the other hand,in Confindence interval ,We calculate a point estimate of population parameter from a sample statistic.If the estimator is an unbiased estimator, then the population value estimated using point estimate will be near the sample statistic. So,we try to include some values lying near the sample statistic value ,so that we can be confident that our population value lies in a range.

In hypothesis testing, the significance level graph is centered around the null hypothesis value while in confidence interval,the same is centered around the sample statistic.

**Similarity-**

Confidence interval and Hypothesis tests both are statistical inference techniques in which we try to estimate population parameter based on sample statistic.

Both use a P-value and a significance level(alpha) to determine weather the results are correct or not.

Results from both will be same, if we assume a null hypothesis that sample statistic is  equal to population parameter.

Type 1 error refers to the error when we reject the null hypothesis although it is correct, because of a rare sample.For ex-A diciplined student is punished.

Type 2 error refers to a condition when we fail to reject a null hypothesid when it is wrong.For ex-An undisciplined student is made monitor.

Support vector regression(SVR) is a continuation of support vector classification(SVC).They both differ in the loss function and constraint.

For SVM classification the hinge loss is used, for SVM regression the epsilon insensitive loss function is used.

## 2. When SVM is not a good approach?

1)When our data is suffering from curse of dimensionality i.e number of training points are less as compared to number of features.

2)SVM becomes very slow when we lots of data points.As it uses kernel trick to project data into other dimensions to find the best hyperplane,It is computationally complex.

Learning rate is used in gradient descent algorithm which is used to minimize the loss function .We update the weights in neural network after every iteration until we reach a minimum value.It is a hypermarameter which needs to be tuned .How quickly weights will get updated is determined by learning rate.If our model as has very low learning rate ,it will take a lot of iterations(and number of epochs that is how many times model passes through whole training dat) to reach the minima.If learning  rate is high,Our model might oscillate around minima and miss the minima even after reaching near it earlier as compared to small learning rate.

So, what we do is we use learning rate decay method where learning rate is large in beginning it decays with increase in iterartions.This way we can converge to minima faster and won't miss it when we reach near minima.

## 4. What are the different types of deep neural networks?

**Different types of deep neural networks-**

**1) Multilayer perceptrons-** This is most simple type of DNN.This is series of fully connected layers.Every node has non-linearity applied to the weighted sum of output of the previous layers.This can learn complex data patterns ehich conventional ML algorithms aren't able to.

**2)Convolution neural networks-** This is used when we need to classify images or image related tasks such as face recognition,computer vision.Different layers learns different features of images .Filters and pooling layers are used for this purpose.

**3)Recurrent Neural network-** RNN's are used when sequential data is available.For audio,video, NLP,time series forecasting.The input of RNN consists of the current input and the previous samples.
