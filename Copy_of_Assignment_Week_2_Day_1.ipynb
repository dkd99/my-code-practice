## 1. Why is Logistic Regression termed as Regression and not classification?

If we look at the logistic regression in terms of a simple neural network,it can be divided into two parts in sequence-

1)A linear combination of input features and their respective weights

2)A non linear activation on after linear combination

The first part i.e linear combination is similar to what we do in linear regression .Logistic regression uses the same basic formula as linear regression but it is regressing for the probability of a categorical outcome.


image.png

So, till the 1st part Logistic regression is same as linear regression.But since logistic regression is used for classification problem we use a activation function which is called sigmoid function to bring the linear summation between 0 and 1.The value near 0 is classified as class 0 and value near class 1 is classified as class 1.

Similar to linear regression, in logistic regression also we try to estimeate the bias and weight terms that minimize the cost function.

## 2. Why can’t we use Mean Square Error (MSE) as a cost function for Logistic Regression?

We choose cost function in such a way that if there is some mismatch in predicted value and original value then our cost function should give a very high value whereas if the predicted and actual values are same cost function should give a value close to 0.

In logistic regeression, our output value is either 0 or 1.If our model misclassifies output it will give 0 for 1 and 1 for 0.If we calculate MSE for this error it will be (0-1)^2,which is almost equal to zero.So,mse isn't able to penalize our model enough for misclassification.To overcome this problem we use binary cross entrpy loss function which gives very high loss value on misclassification as it has log terms in it.And log(0) is infinite value.

## 3. What do you understand by Pruning in a Decision Tree and Why do we require Pruning in Decision Trees? Explain.

Decision tree learns by partitioning the feature space into regions and then using these regions to predict the dependent variable.When complete growth of decision tree is allowed while training ,there is a high chance that it will give good performance on train data but not so good on unseen test data.This is beacause decision tree is prone to overfitting.Decision trees grow untill the leaves don't become homogeneous i.e of single class.

So, to prevent overfitting we prune the decision tree i.e trim parts of it which do not provide the power to classify instances.

There are two ways to prune a tree-
1)Pre-prunning-This doesn't lets the decision tree to grow fully by using some hyperparameters like max_depth, min_samples_leaf, and min_samples_split .

2)Post-prunning- This method allows the tree to grow fully.Then it uses the cost complexity formula i.e we remove some leave and calculate tree score based on formula.Formula is  

Tree score= SSE+alpha *(number of leaves)
Here,alpha is the hyperparameter we need to tune so that we could get low test error and our model doesn't overfits.


## 4. What do you understand about Information Gain? Also, explain the mathematical formulation associated with it.


Information gain refers to weather the information in nodes improved on slitting or not.Decision tree splits based on certain feature trying to achieve homogenity in further nodes i.e the nodes after splitting contains one type of class.So ,information gain tells us that whether the split based on feature is able to achieve this homogenity or not.If after splitting an impure node(heterogenious node) on a certain feature we are able to get nearly pure nodes then we can say we have gained more information as compared to split based on some other feature which wasn't able to achieve that much homegenity.
