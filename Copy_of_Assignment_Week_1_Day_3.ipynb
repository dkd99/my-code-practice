## 1. What is multicollinearity, and how do you treat it and when is it not a problem?

Multicollinearity occurs when there is dependency or association between independent variables.The coefficient corresponding to independent variable in regression analysis represents the change in dependent variable w.r.t independent variable when all other independent variables are kept constant.But when there is correlation between one or more independent variables,changes in one variable are associated with shifts in another variable. The stronger the correlation, the more difficult it is to change one variable without changing another.

**When multicollinearity isn't a problem-**
1)When there is very less multicollinearity then there is no need to resolve it as it won't affect the coefficients and prediction much.

2)If multicollinearity isn't present for variables that we are interested in then there isn't a need to resolve it.

3) If our primary goal is to make predictions, and we don’t need to understand the role of each independent variable, we don’t need to reduce severe multicollinearity since multicollinearity doesn't affect predictions.

**How to treat Multicollinearity-**
1)Remove some of the highly correlated independent variables.

2)Linearly combine the independent variables, such as adding them together.

3)Using PCA to reduce the number of variables.PCA will create new variables known as Principal components that are uncorrelated. 

4)LASSO and Ridge regression are advanced forms of regression analysis that can handle multicollinearity.We can use these instead of linear regression.

## 2. VIF(Variation Inflation Factor),Weight of Evidence & Information Value. Why and when to use?

**A variance inflation factor (VIF)** is a measure of the amount of multicollinearity in regression analysis. 

pairwise correlation between all predictors in the model can be used to assess the degree of correlation. However, There  may be a scenario when we have five predictors and the pairwise correlations between each pair are not exceptionally high and it is still possible that three predictors together could explain a very high proportion of the variance in the fourth predictor.

                                 VIF = 1 /(1- Rj^2)
Rj2 is the R squared of the model of one individual predictor against all the other predictors. 

If VIF <5,  this particular variable cann't be explained using other variables
and if VIF>5 , this particular variable can be explained using other variables and needs treatment.


** Weight of Evidence and Information value-  **



These techniques have evolved from logistic regression technique used for binary classification.These techniques have evolved from credit scoring world.

For numerical features,there’s a natural ordering for various numerical values but nonlinearity may exist which is difficult to be  captured by linear model.Similarly  categorical features are converted into vector form using one hot encoding but The 0-1 (binary) valued vectors are tough to interpret as a feature.So to overcome these problems we use WOE.

**The weight of evidence** 

Weight of Evidence (WOE) helps to transform a continuous independent variable into a set of groups or bins based on similarity of dependent variable distribution i.e. number of events and non-events.The benefit of converting continous variables to a set of bins is that this makes model more stable as small change in variable doesn't impact the model.We can also transform categorical variables to a numeric varaible using woe as it has certain benefits over one hot encoding and label based encoding.Calculated WOE values of different classes can be used to replace their actual values.The number of bins should be in such a way that WOE values follows visible particular order.

**Information value-** 

A numerical value that quantifies the predictive power of an independent continuous variable x in capturing the binary dependent variable y. IV is helpful for reducing the number of variables as an initial step in preparing for Logistic Regression, especially when there are a large amount of potential variables. IV is based on an analysis of each individual independent variable in turn without considering other predictor variables.

## 3. What is tradeoff between bias and variances, and the relationship between them?

An ideal model should have low train as well as low test error.

A model has high Bias when the fitted curve doesn't changes much when a new observation is added .On the other hand, If a model is having high variance,the curve will change a lot in order to pass through a newly added point.A highly biased model model is too lazy to move to capture the pattern in the data while a model having a very high variance is so active that instead of capturing pattern ,it ends up capturing noise in the data .

Weather a model will have high bias or variance is determined by the complexity of model.

A biased  model is quite simple and  has both high train error as well as test error .A high variance model is quite complex and has low training error but high test error.So, we can see there is an inverse relationship between bias and variance

So, an ideal model should have neither high bias nor high variance .There should be a tradeoff between both .

## 4. How overfitting and underfitting can affect model generalization?



Generalization is a term used to describe a model’s ability to react to new data.
Overfitting implies that model is performing well on train data but not on test data.It will make inaccurate predictions when given new data, making the model useless even though it is able to make accurate predictions for the training data i.e model is unable to generalize.

Underfitting implies that model isn't capturing the underlying pattern in data.Both test and train error are high in an underfitting model.The model knows that there is some trend in the data, but it is not able to capture relevant information.

There will be a sweet spot between underfitting and overfitting complexity which will be able to generalize on the pattern present in data and perfrom well on both test and training data.
