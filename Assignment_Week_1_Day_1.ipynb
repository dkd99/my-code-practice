## 5. How do you choose a machine learning algorithm for a problem statement?

It isn't considered  best practice to use and evaluate every machine learning algorithm to find the best one .Here are few steps I use to choose a machine learning alogorithms for a given problem statement-

**1) Type of problem-**
There are different type of machine learning algorithms.Supervised machine learning algorithms is given labelled data (one column), and  the remaining data (rest of columns), utilized as input features as information.Supervised learning is further subdivided into regression as well as classification modules.Unsupervised learning is given only  input features  based on which it is able to analyse the patterns in data  that are not immediately apparent to the human eyes.And then there is reinforcement learning which works similar to the way human learns.

Once you identify the type of problem you are working with, you can finally make a choice of algorithm based on the type of problem.If we want to make future prediction based on past data we can go with forecasting.If the problem is image classification then CNN is the best choice.

**2)Type of data-**
If our input features are unstructured like images,audio and video then its better to use deep learning algorithms.Also if we have a lot of training data available its better to use deep learning algorithms because it works well when we have lot of data available.

**3)Accuracy and interpretability-**
If low accuracy of around 80% is required then its better to use simple and explainable models as compared to complex models.Also, we won't require to do more  hyper parameter tuning to improve accuracy. 

**4)Training time required-**
For applications like movies recommendation data needs to be trained every time a user logs in while in some applications like stock predictions model needs to be trained every second.Deep neaural networks require a large training time while models like linear ,logistic regressions and KNN require less training time.

**5)Number of features-**
Sometimes number  of features might be very large as compared to datapoints .Some algorithms like SVM  don't perform well when this happens and training becomes infeasible.Then its better to use Deep neural networks .We may use dimensionilty reduction algorithms like PCA to improve our model performance.

**6)Linear or not-**
If all the components the components of the feature vector are statistically independent. Then covariance matrix that we get will be a diagonal
matrix and if every component has same variance,then we can be sure that the decison boundary that seperates the two classes will be a linear boundary.If these conditions are not satisfied then deep learning or ensemble learning will be a better choice.

**7)Size of training set- ** 
When size of training set is low then its better to use high bias/low variance models and as training data increases its better to use high variance/low bias algorithms as low bias ones aren't able to predict patterns in data.We can use reguralization to prevent overfitting .

**8)Missing values**
If there are lot of missing values then we should use algorithms which are immune to missing values. Ensemble models or generative models perform well when there are missing values.

## 7. Why we can’t do a classification problem using Regression?

Classification problem is focussed on predicting label while Regression is used for predicting contionous values.The output of regression can range between -inf to + inf.In classification,outputs are probabilities ranging between 0 and 1.We set a thresold value.If probability is above thresold it can be classified as 1 and if probablity is below thresold it will be classified as 1.

We can train linear regression model on training data having output values  for classes as 0 and 1.But this doesn't gaurantee that the output for the test data will also lie between 0 and 1.It can go above 1 and below zero as well which is not what we want as probabilities cann't be greater than 1 or less than 0.

As we know, linear regression is highly affected by outliers.Everytime an oulier is added to the data the regression line  which we have fitted before shifts significantly changing the prediction. To compensate  this change in prediction, we have to change the thresold value but that is  not how the algorithm should work. 

Loss function, in case of  regression is mean squared error which doesn't gives a large value when we misclassify class 0 as 1 or vice versa.The best loss function for classification problem is binary cross entropy which gives a huge loss on misclassification.

Using linear regression for multi class prediction makes no sense. Linear regression assumes an order between 0, 1, and 2, whereas in the classification regime these numbers are mere categorical placeholders.

We need an algorithm that absorbs the effects of outliers without impacting the final output. Logistic regression does that by using something called a Sigmoid function.For multicass problems we use softmax function .


## 8. How do you treat heteroscedasticity in regression? What causes Heteroscedasticity and how to fix it?

Heteroscadity refers to a situation  where the variance of the residuals is unequal over a range of measured values.Heteroscedasticity does not cause bias in the coefficient estimates, it does make them less precise.

Causes of heteroscadity-

1)If dataset contains a wide range of values then there are high chances of having heteroscadity.

2)Heteroscedasticity can also arise as a result of the presence of outliers. The inclusion or exclusion of such observations, especially when the sample size is small, can substantially alter the results of regression analysis.

3)Skewness in the distribution of one or more regressors included in the model is another source of heteroscedasticity.


Fixing heteroscadity-

1) Redefining the variables-This means instead of using raw values use proportion so that the scaling problem is removed.For ex-Instead of using population of city to predict number of accidents we can predict rate of accident happening.

2) Transforming the data- We can use log transformation or box cox transformation of features . log tranformation reduces the gap between largest and smallest values.For ex-log(10^4) is 4 while log(10) is 1.

3)We can also treat heteroscadity by removing outliers.

4)Use weighted regression-This type of regression assigns a weight to each data point based on the variance of its fitted value.Essentially, this gives small weights to data points that have higher variances, which shrinks their squared residuals. When the proper weights are used, this can eliminate the problem of heteroscedasticity.
